````python
def get_completion(batch, current_batch, total_batch, model='deepseek-chat'):
    print(f"Processing batch {current_batch+1} of {total_batch}")
    
    # Number sentences so the model matches outputs in order
    sentences_text = "\n".join(
        [f"ID: {idx} | {s}" for idx, s in zip(batch['id'], batch['sentence'])]
    )
    
    zero_shot_prompt = f"""
You are a precision aspect-based sentiment analysis system specialized in Meta's business domains.

TASK: Analyze each sentence for sentiment toward Meta's 6 core business aspects. - The sentences are provided between three backticks below.

ASPECT DEFINITIONS:
{categories}

KEYWORD INDICATORS:
{key_words}

SENTIMENT SCORING RULES:
- **-1 (Negative)**: Problems, criticism, declines, challenges, concerns, failures
- **0 (Neutral)**: Factual statements, balanced reporting, mixed signals, announcements
- **1 (Positive)**: Growth, success, improvements, opportunities, praise, achievements  
- **2 (Not Relevant)**: Aspect not mentioned, discussed, or contextually implied

ANALYSIS PROTOCOL:
1. **Identify relevant aspects** - Use keywords as indicators but consider broader context
2. **Determine sentiment direction** - Focus on the overall tone toward each aspect
3. **Handle multi-aspect sentences** - Score each aspect independently based on its specific context

CRITICAL OUTPUT REQUIREMENTS:
- Use **integer scores only**: -1, 0, 1, or 2
- Return **only** a valid JSON array** where each element corresponds to one sentence, in the same order as listed.
- Each element must follow exactly this structure: {json.dumps(output_format, indent=2)}.


    ```
    {sentences_text}
    
    ```
    """
    
    print(zero_shot_prompt)
    messages=[{"role": "user", "content": zero_shot_prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": zero_shot_prompt}],
        stream=False
    )
    return response.choices[0].message.content
````

````python
def get_completion(batch, current_batch, total_batch, model='gpt-5'):
    print(f"Processing batch {current_batch+1} of {total_batch}")
    
    # Number sentences so the model matches outputs in order
    sentences_text = "\n".join(
        [f"ID: {idx} | {s}" for idx, s in zip(batch['id'], batch['sentence'])]
    )
    
    zero_shot_prompt = f"""
You are a precision aspect-based sentiment analysis system specialized in Meta's business domains.

TASK: Analyze each sentence for sentiment toward Meta's 6 core business aspects. - The sentences are provided between three backticks below.

ASPECT DEFINITIONS:
{categories}

KEYWORD INDICATORS:
{key_words}

SENTIMENT SCORING RULES:
- **-1 (Negative)**: Problems, criticism, declines, challenges, concerns, failures
- **0 (Neutral)**: Factual statements, balanced reporting, mixed signals, announcements
- **1 (Positive)**: Growth, success, improvements, opportunities, praise, achievements  
- **2 (Not Relevant)**: Aspect not mentioned, discussed, or contextually implied

ANALYSIS PROTOCOL:
1. **Identify relevant aspects** - Use keywords as indicators but consider broader context
2. **Determine sentiment direction** - Focus on the overall tone toward each aspect
3. **Handle multi-aspect sentences** - Score each aspect independently based on its specific context

CRITICAL OUTPUT REQUIREMENTS:
- Use **integer scores only**: -1, 0, 1, or 2
- Return **only** a valid JSON array** where each element corresponds to one sentence, in the same order as listed.
- Each element must follow exactly this structure: {json.dumps(output_format, indent=2)}.


    ```
    {sentences_text}
    
    ```
    """
    
    print(zero_shot_prompt)
    messages=[{"role": "user", "content": zero_shot_prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": zero_shot_prompt}],
    )
    return response.choices[0].message.content
````

````python
batch_count = len(batches)
responses = []

for i in range(0,len(batches)):
    response = get_completion(batches[i], i, batch_count)
    responses.append(response)
````

````python
import json
import re
import pandas as pd

df_total_gpt = pd.DataFrame()

for i, response in enumerate(responses): 
    try:
        # Clean code block markers if present
        cleaned = re.sub(r"^```(json)?|```$", "", response.strip()).strip()
        data = json.loads(cleaned)
        df_temp = pd.DataFrame(data)
        df_total_gpt = pd.concat([df_total_gpt, df_temp], ignore_index=True)
    except json.JSONDecodeError as e:
        print(f"JSON decode error in batch {i+1}: {e}")
        print("Raw response:")
        print(response)
        
df_total_gpt
````
